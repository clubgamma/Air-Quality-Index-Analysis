```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Air-Quality-Index-Analysis/data/city_day.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
city_aqi = data.groupby('City')['AQI'].mean() ## calculating the mean of each city
city_aqi
```

```python
## bar graph representation of mean of AQI of different cities
plt.figure(figsize=(10, 6))
city_aqi.plot(kind='bar', color='skyblue')
plt.title('Average AQI Levels by City')
plt.xlabel('City')
plt.ylabel('Average AQI')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
## Line chart representation of the AQI in different cities
plt.figure(figsize=(12, 6))
plt.plot(data['City'], data['AQI'], marker='o', linestyle='-', color='blue')
plt.title('Proxy AQI Levels by City (Line Chart)')
plt.xlabel('City')
plt.ylabel('Proxy AQI')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['City'], data['AQI'], color='purple')
plt.title('Proxy AQI Levels by City (Scatter Plot)')
plt.xlabel('City')
plt.ylabel('Proxy AQI')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['AQI'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Proxy AQI Levels Across Cities')
plt.xlabel('Proxy AQI')
plt.ylabel('Number of Cities')
plt.tight_layout()
plt.show()
```
